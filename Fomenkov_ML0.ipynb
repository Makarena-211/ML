```python
#Создайте функцию, которая принимает как аргументы целое число N
#и первый элемент (вещественное число), и разность (вещественное число) и создает матрицу numpy по диагонали,
#которой располагаются первые N членов арифметической прогрессии.

def profdiag(N, first, delta):
    d = np.arange(first, first+N*delta, delta)
    return np.diag(d)
print(profdiag(4, 1.28, 3.02))
```

```python
#Удалите из таблицы столбцы, содержащие идентификаторы,
#переименуйте все оставшиеся названия колонок на русском языке.

df=pd.read_csv('data.csv', index_col = "Id")
A=['Id','DistrictId','Rooms','Square','LifeSquare','KitchenSquare','Floor','HouseFloor','HouseYear','Ecology_1','Ecology_2','Ecology_3','Social_1','Social_2','Social_3','Healthcare_1','Helthcare_2','Shops_1','Shops_2','Price']
B=['Идентификатор', 'Районный номер', 'Комнаты', 'Площадь', 'Площадь_жизни', 'Площадь_кухни', 'Этаж', 'Этаж_дома', 'Год_дома', 'Экология_1', 'Экология_2', 'Экология_3', 'Социальная_1', 'Социальная_2', 'Социальная_3', 'Здравоохранение_1', 'Медицинская_2', 'Магазины_1', 'Магазины_2', 'Цена']
df9=df.rename(columns = dict(zip(A, B)))
print(df9)
```

```python
#Выведите отдельно столбец, содержащий цену, по номеру и названию.
#Выведите первую, десятую и предпоследнюю строку таблицы по номеру и по индексу.

print(df9['Цена'])
print(df9.iloc[:,-1])
print(df9.iloc[[0,9,-2], -1])
print(df9.loc[[11809, 8094, 14561], "Цена"])
```

```python
#Выделите в отдельную таблицу последние десять строк.
#Уберите в ней столбец с ценой. Склейте ее с первоначальной таблицей при помощи append.
#Заполните отсутствующие значения цены средним по таблице.

df11 = df9.tail(10)
df11=df11.drop("Цена", axis=1)
print(df11)
df11 = df9.append(df11)
print(df11)
```

```python
#Выделите пять последних колонок в отдельную таблицу. Удалите в ней строки, в которых цена ниже среднего.
#Присоедините эту таблицу к изначальной (выберите самый подходящий тип соединения).

df12 = df9.iloc[:,-5:]
df12 = df12[df12["Цена"] >= df12["Цена"].mean()]
df12 = pd.concat([df9, df12])
print(df12)
```

```python
#Выведите таблицу,содержащую среднюю цену и количество квартир на каждом этаже из первоначального набора данных.

print(df9.groupby("Этаж")['Цена'].count())
print(df9.groupby("Этаж")["Цена"].aggregate(['mean','count']))
```

```python
df_5 = df[["Rooms", "Square", "HouseFloor", "HouseYear", "Price"]]
g = sb.PairGrid(df_5)
g.map_diag(plt.hist)
g.map_upper(plt.scatter)
g.map_lower(sb.kdeplot)
```
